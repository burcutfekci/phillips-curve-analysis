```python
import pandas as pd
import requests
import json

def wb_get(country, indicator):
    url = f"https://api.worldbank.org/v2/country/{country}/indicator/{indicator}?format=json&per_page=5000"

    response = requests.get(url)

    # Eğer World Bank API HTML döndürürse JSON decode hatası çıkmasın:
    try:
        data = response.json()
    except:
        print("API JSON döndürmedi. Bağlantı yenileniyor...")
        response = requests.get(url)
        data = response.json()

    # Asıl veri her zaman 2. elemanda (index 1)
    if len(data) < 2 or data[1] is None:
        raise ValueError("API'den veri gelmedi. Lütfen tekrar çalıştır.")

    df = pd.json_normalize(data[1])
    df = df[['date', 'value']]
    return df

# Türkiye İşsizlik
unemp = wb_get("TUR", "SL.UEM.TOTL.ZS")
unemp = unemp.rename(columns={'value': 'unemployment'})

# Türkiye Enflasyon (TÜFE)
infl = wb_get("TUR", "FP.CPI.TOTL.ZG")
infl = infl.rename(columns={'value': 'inflation'})

# Birleştir
df = pd.merge(unemp, infl, on="date")
df = df.sort_values("date", ascending=True)
df = df.dropna()

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# sütun isimlerini normalize et
df.columns = df.columns.str.lower().str.strip()

# unemployment = işsizlik sütunu
unemp_col = [c for c in df.columns if "unemp" in c][0]

# inflation = enflasyon sütunu
infl_col = [c for c in df.columns if "infl" in c][0]

plt.figure(figsize=(8,6))
sns.regplot(x=df[unemp_col], y=df[infl_col], scatter_kws={'s':40})
plt.xlabel("Unemployment (%)")
plt.ylabel("Inflation (%)")
plt.title("Phillips Curve - Turkey")
plt.show()
```
